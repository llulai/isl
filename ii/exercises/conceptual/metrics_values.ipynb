```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
variables = {
    'bias' : lambda x: x **2 - 20 * x,
    'train error' : lambda x: x ** 2 - 20 * x + 15,
    'variance' : lambda x: x ** 2 - 100,
    'test error': lambda x: x**2 - 20*x + x**2+ 30,
    'bayes error': lambda x: -40
}
```

```python
X = np.linspace(0.1, 10, 100)
```

```python
variables_y = {var_name: [fn(x) for x in X] for var_name, fn in variables.items()}
```

```python
plt.figure(figsize=(15, 10))
for var_name, var_y in variables_y.items():
    plt.plot(X, var_y, label=var_name)

plt.legend(loc='upper center', fontsize=14)
plt.xticks([])
plt.yticks([])
plt.xlabel('flexibility', fontsize=20)
plt.ylabel('metric value', fontsize=20)
plt.show()
```
